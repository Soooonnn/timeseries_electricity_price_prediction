```python
df['hour'] = pd.to_datetime(df['hour'])

# hour별 평균 smp_rc 계산
hourly_data = df.groupby('hour')['smp_rc'].mean().reset_index()

# 결과 확인
print(hourly_data)
```

```python
import pandas as pd

# 데이터프레임에서 시간대별 IQR 계산 및 Outlier 탐지
def detect_outliers_iqr_by_hour(data, k):
    # 시간별로 IQR 계산
    iqr_stats = data.groupby(data['hour'].dt.hour)['smp_rc'].quantile([0.25, 0.75]).unstack()
    iqr_stats['IQR'] = iqr_stats[0.75] - iqr_stats[0.25]
    iqr_stats['P_TH'] = iqr_stats[0.75] + k * iqr_stats['IQR']
    iqr_stats['P_TL'] = iqr_stats[0.25] - k * iqr_stats['IQR']

    # 시간대별 IQR 기준으로 Outlier 탐지
    data['hour_only'] = data['hour'].dt.hour
    data = data.merge(iqr_stats[['P_TH', 'P_TL']], left_on='hour_only', right_index=True)
    data['positive_outlier'] = data['smp_rc'] > data['P_TH']
    data['negative_outlier'] = data['smp_rc'] < data['P_TL']

    return data

# 실행
hourly_data['hour'] = pd.to_datetime(hourly_data['hour'])  # ensure 'hour' is datetime
result = detect_outliers_iqr_by_hour(hourly_data, 1.5)

# 결과 확인
print("시간대별 Outlier 탐지 결과:")
print(result[['hour', 'smp_rc', 'P_TH', 'P_TL', 'positive_outlier', 'negative_outlier']])

result.to_csv('1_5IQR.csv')
```

### z-score

```python
# 시간대별 Z-Score 기반 Outlier 탐지
def detect_outliers_zscore_by_hour(data, threshold=2):
    # 시간대별 평균과 표준편차 계산
    group_stats = data.groupby(data['hour'].dt.hour)['smp_rc'].agg(['mean', 'std']).reset_index()
    group_stats.rename(columns={'mean': 'mean_smp_rc', 'std': 'std_smp_rc', 'hour': 'hour_only'}, inplace=True)

    # 시간대별 기준값 병합
    data['hour_only'] = data['hour'].dt.hour
    data = data.merge(group_stats, left_on='hour_only', right_on='hour_only')

    # Z-Score 계산 및 Outlier 판단
    data['z_score'] = (data['smp_rc'] - data['mean_smp_rc']) / data['std_smp_rc']
    data['outlier_type'] = 'normal'
    data.loc[data['z_score'] > threshold, 'outlier_type'] = 'positive'
    data.loc[data['z_score'] < -threshold, 'outlier_type'] = 'negative'

    return data

# 실행
result = detect_outliers_zscore_by_hour(hourly_data)
print(result[['hour', 'smp_rc', 'z_score', 'outlier_type']])
```

```python
# 시간대별 Z-Score 기반 시각화
def plot_zscore_outliers(data):
    plt.figure(figsize=(12, 6))
    
    # 전체 데이터 플롯
    plt.plot(data['hour'], data['smp_rc'], label='smp_rc', color='blue', alpha=0.7)
    
    # Positive Outliers
    positive_outliers = data[data['outlier_type'] == 'positive']
    plt.scatter(positive_outliers['hour'], positive_outliers['smp_rc'], color='red', label='Positive Outliers')
    
    # Negative Outliers
    negative_outliers = data[data['outlier_type'] == 'negative']
    plt.scatter(negative_outliers['hour'], negative_outliers['smp_rc'], color='green', label='Negative Outliers')
    
    # 평균선 추가 (각 시간대의 평균)
    for hour, row in data.groupby('hour_only').first().iterrows():
        plt.axhline(row['mean_smp_rc'], color='orange', linestyle='--', alpha=0.5, label='Mean (Hour {})'.format(hour) if hour == 0 else "")

    plt.xlabel('Time')
    plt.ylabel('smp_rc')
    plt.title('Time-Based Z-Score Outlier Detection')
    plt.legend()
    plt.show()

# 시각화 실행
plot_zscore_outliers(result)
```

```python
# 시간대별 IQR 기반 Outlier 탐지
def detect_outliers_iqr_by_hour(data, k=1.5):
    # 시간대별 IQR 계산
    group_stats = data.groupby(data['hour'].dt.hour)['smp_rc'].quantile([0.25, 0.75]).unstack()
    group_stats['IQR'] = group_stats[0.75] - group_stats[0.25]

    # 상한(P_TH)과 하한(P_TL) 계산
    group_stats['P_TH'] = group_stats[0.75] + k * group_stats['IQR']
    group_stats['P_TL'] = group_stats[0.25] - k * group_stats['IQR']

    # 시간대별 기준값 병합
    data['hour_only'] = data['hour'].dt.hour  # 시간(Hour) 추출
    data = data.merge(group_stats[['P_TH', 'P_TL']], left_on='hour_only', right_index=True)

    # Outlier 판단 및 구분
    data['outlier_type'] = 'normal'
    data.loc[data['smp_rc'] > data['P_TH'], 'outlier_type'] = 'positive'
    data.loc[data['smp_rc'] < data['P_TL'], 'outlier_type'] = 'negative'

    return data

# 실행
result = detect_outliers_iqr_by_hour(hourly_data)
print(result[['hour', 'smp_rc', 'outlier_type']])
```

### Isolation Forest

```python
from sklearn.ensemble import IsolationForest
import pandas as pd

# Isolation Forest로 시간대별 이상치 탐지
def detect_outliers_isolation_forest_by_hour(data, contamination=0.05):
    # 시간대별로 처리
    outlier_results = []

    for hour, group in data.groupby(data['hour'].dt.hour):
        model = IsolationForest(contamination=contamination, random_state=42)
        group['outlier_type'] = model.fit_predict(group[['smp_rc']])  # -1: 이상치, 1: 정상
        group['outlier_type'] = group['outlier_type'].replace({1: 'normal', -1: 'outlier'})
        outlier_results.append(group)

    # 모든 시간대 결과 병합
    return pd.concat(outlier_results)

# 실행

result = detect_outliers_isolation_forest_by_hour(hourly_data)

# 결과 확인
print(result[['hour', 'smp_rc', 'outlier_type']])
```

```python
def analyze_outliers(data):
    # 시간대별 평균 계산
    hour_stats = data.groupby(data['hour'].dt.hour)['smp_rc'].mean().reset_index()
    hour_stats.rename(columns={'smp_rc': 'mean_smp_rc', 'hour': 'hour_only'}, inplace=True)
    
    # 데이터에 시간대별 평균 병합
    data['hour_only'] = data['hour'].dt.hour
    data = data.merge(hour_stats, left_on='hour_only', right_on='hour_only')
    
    # Outlier 타입 구분
    data['positive_outlier'] = (data['outlier_type'] == 'outlier') & (data['smp_rc'] > data['mean_smp_rc'])
    data['negative_outlier'] = (data['outlier_type'] == 'outlier') & (data['smp_rc'] < data['mean_smp_rc'])
    
    # 시간대별 Positive와 Negative Outlier 개수 계산
    outlier_summary = data.groupby('hour_only').agg({
        'positive_outlier': 'sum',
        'negative_outlier': 'sum'
    }).reset_index()

    outlier_summary.rename(columns={
        'positive_outlier': 'positive_outlier_count',
        'negative_outlier': 'negative_outlier_count'
    }, inplace=True)

    return outlier_summary, data

# 실행
outlier_summary, updated_data = analyze_outliers(result)

# 결과 확인
print("시간대별 Outlier Summary:")
print(outlier_summary)
```

### LOF

```python
from sklearn.neighbors import LocalOutlierFactor
import pandas as pd

# Local Outlier Factor로 시간대별 이상치 탐지
def detect_outliers_lof_by_hour(data, n_neighbors=20):
    outlier_results = []

    for hour, group in data.groupby(data['hour'].dt.hour):
        # LOF 모델 학습
        lof_model = LocalOutlierFactor(n_neighbors=n_neighbors, contamination=0.05, novelty=False)
        group['lof_score'] = lof_model.fit_predict(group[['smp_rc']])  # -1: 이상치, 1: 정상

        # Positive/Negative Outlier 구분
        mean_value = group['smp_rc'].mean()  # 시간대별 평균
        group['outlier_direction'] = 'normal'
        group.loc[(group['lof_score'] == -1) & (group['smp_rc'] > mean_value), 'outlier_direction'] = 'positive'
        group.loc[(group['lof_score'] == -1) & (group['smp_rc'] < mean_value), 'outlier_direction'] = 'negative'

        outlier_results.append(group)

    # 모든 시간대 결과 병합
    return pd.concat(outlier_results)

# 실행
result = detect_outliers_lof_by_hour(hourly_data)

# 결과 확인
print(result[['hour', 'smp_rc', 'lof_score', 'outlier_direction']])
```

```python
# smp_rc와 뒤에서 10개를 제외한 컬럼 간 상관관계 분석
def analyze_correlations(df, target_column='smp_rc', exclude_last_n=10):
    # 뒤에서 10개 컬럼 제외
    selected_columns = df.columns[:-exclude_last_n]  # 뒤에서 exclude_last_n 개 컬럼 제외
    selected_columns = [col for col in selected_columns if col != target_column]  # target_column 제외
    
    # 상관관계 계산
    correlations = df[selected_columns].corrwith(df[target_column])
    
    # 결과 정렬
    correlations = correlations.sort_values(ascending=False).reset_index()
    correlations.columns = ['Feature', 'Correlation with smp_rc']
    
    return correlations

# 실행
correlation_results = analyze_correlations(df2, target_column='smp_rc', exclude_last_n=10)

# 결과 출력
print("smp_rc와 각 컬럼 간 상관관계:")
print(correlation_results)
```
